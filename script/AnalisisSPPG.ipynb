```python
from shapely.geometry import Point

#10 Konversi excel ke geodataframe
def sekolah_to_gdf(df, jenjang):
    df = df.copy()

# Buat kolom jenjang yang nilainya diambil dari jenjang
    df["jenjang"] = jenjang

# Ngubah data ke numeric
    df["lon"] = pd.to_numeric(df["lon"], errors="coerce")
    df["lat"] = pd.to_numeric(df["lat"], errors="coerce")
    df["jumlah_siswa"] = pd.to_numeric(df["jumlah_siswa"], errors="coerce")

    df = df.dropna(subset=["lon", "lat"])

    geometry = [Point(xy) for xy in zip(df["lon"], df["lat"])]
                
                
    gdf = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")

    return gdf

sd_gdf = sekolah_to_gdf(sd, "SD")
smp_gdf = sekolah_to_gdf(smp, "SMP")
sma_gdf = sekolah_to_gdf(sma, "SMA")

print("SD :", len(sd_gdf))
print("SMP:", len(smp_gdf))
print("SMA:", len(sma_gdf))


   

```

```python
# ============================================================
# 16. MEMBUAT MULTI-RING BUFFER MURNI SPPG EXISTING
# ============================================================
# Tujuan:
# Membuat zona layanan SPPG:
# 0-1 km = sangat dekat
# 1-2 km = sedang
# 2-3 km = batas maksimal layanan
# >3 km = gap layanan

# Dissolve semua buffer agar overlap antar-SPPG menyatu
buffer_1km_union = sppg_buffer_1km.dissolve()
buffer_2km_union = sppg_buffer_2km.dissolve()
buffer_3km_union = sppg_buffer_3km.dissolve()

# Membuat ring murni
ring_0_1km = buffer_1km_union.copy()
ring_0_1km["kelas_ring"] = "0-1 km"
ring_0_1km["skor_gap"] = 1

ring_1_2km = buffer_2km_union.copy()
ring_1_2km["geometry"] = buffer_2km_union.geometry.difference(buffer_1km_union.geometry.iloc[0])
ring_1_2km["kelas_ring"] = "1-2 km"
ring_1_2km["skor_gap"] = 2

ring_2_3km = buffer_3km_union.copy()
ring_2_3km["geometry"] = buffer_3km_union.geometry.difference(buffer_2km_union.geometry.iloc[0])
ring_2_3km["kelas_ring"] = "2-3 km"
ring_2_3km["skor_gap"] = 3

# Area di luar 3 km belum dibuat polygon-nya.
# Nanti gap utama dihitung dari kandidat/sekolah yang tidak masuk buffer 3 km.
```

```python
# ============================================================
# 17. GABUNG MULTI-RING BUFFER
# ============================================================

multi_ring_sppg = pd.concat(
    [ring_0_1km, ring_1_2km, ring_2_3km],
    ignore_index=True
)

multi_ring_sppg = gpd.GeoDataFrame(
    multi_ring_sppg,
    geometry="geometry",
    crs=TARGET_CRS
)

display(multi_ring_sppg[["kelas_ring", "skor_gap", "geometry"]])
```

```python
# ============================================================
# C1. PARAMETER JANGKAUAN SPPG EXISTING
# ============================================================
# Logika skor:
# 0–1 km = 1
# 1–2 km = 2
# 2–3 km = 3
# >3 km  = 5

# 1. Poloskan/dissolve batas Surabaya
surabaya_polos = surabaya_m.dissolve()

# 2. Buat zona gap luar 3 km
zona_gap = surabaya_polos.geometry.iloc[0].difference(
    sppg_buffer_3km.dissolve().geometry.iloc[0]
)

zona_gap_gdf = gpd.GeoDataFrame(
    {"kelas_ring": [">3 km"], "skor_c1": [5]},
    geometry=[zona_gap],
    crs=TARGET_CRS
)

# 3. Pastikan multi-ring punya skor C1
multi_ring_c1 = multi_ring_sppg.copy()

multi_ring_c1["skor_c1"] = multi_ring_c1["kelas_ring"].map({
    "0-1 km": 1,
    "1-2 km": 2,
    "2-3 km": 3
})

multi_ring_c1 = multi_ring_c1[["kelas_ring", "skor_c1", "geometry"]]

# 4. Gabungkan ring 0–3 km dan gap >3 km
c1_zona = pd.concat(
    [multi_ring_c1, zona_gap_gdf],
    ignore_index=True
)

c1_zona = gpd.GeoDataFrame(
    c1_zona,
    geometry="geometry",
    crs=TARGET_CRS
)

# 5. Intersect dengan kecamatan
c1 = gpd.overlay(
    surabaya_m[["Kecamatan", "geometry"]],
    c1_zona,
    how="intersection"
)

display(c1[["Kecamatan", "kelas_ring", "skor_c1", "geometry"]].head())
```

```python
# ============================================================
# FISHNET GRID 500m x 500m — UNIT ANALISIS SPASIAL BARU
# ============================================================
# Kecamatan terlalu kasar: seluruh poligon satu warna seragam
# meski sekolah hanya terkonsentrasi di titik tertentu.
# Grid 500m memberikan resolusi spasial yang presisi.

import numpy as np
from shapely.geometry import box, Point
from scipy.spatial import cKDTree
from scipy.stats import gaussian_kde
from matplotlib.lines import Line2D
from matplotlib.colors import LinearSegmentedColormap, ListedColormap
import matplotlib.patches as mpatches

CELL_SIZE = 500  # meter

surabaya_union = surabaya_m.dissolve()
bounds = surabaya_union.total_bounds  # minx, miny, maxx, maxy

xs = np.arange(bounds[0], bounds[2], CELL_SIZE)
ys = np.arange(bounds[1], bounds[3], CELL_SIZE)

grid_cells = [box(x, y, x + CELL_SIZE, y + CELL_SIZE) for x in xs for y in ys]

grid = gpd.GeoDataFrame(geometry=grid_cells, crs=TARGET_CRS)
grid = gpd.clip(grid, surabaya_union)
grid = grid.reset_index(drop=True)
grid["grid_id"] = grid.index
grid["cx"] = grid.geometry.centroid.x
grid["cy"] = grid.geometry.centroid.y

print(f"Total grid sel : {len(grid):,}")
print(f"Resolusi       : {CELL_SIZE}m x {CELL_SIZE}m")
```

```python
# ============================================================
# C1. SPATIAL JOIN JANGKAUAN SPPG RING BUFFER -> GRID
# ============================================================
# c1 (ring polygon) sudah dibuat di sel sebelumnya -> join ke centroid grid

grid_pts = gpd.GeoDataFrame(
    {"grid_id": grid["grid_id"]},
    geometry=gpd.points_from_xy(grid["cx"], grid["cy"]),
    crs=TARGET_CRS
)

c1_join = gpd.sjoin(
    grid_pts,
    c1[["skor_c1", "geometry"]],
    how="left",
    predicate="within"
)

# Jika centroid masuk >1 ring (edge case overlap), ambil max
c1_join = c1_join.groupby("grid_id")["skor_c1"].max().reset_index()
grid = grid.merge(c1_join, on="grid_id", how="left")
grid["skor_c1"] = grid["skor_c1"].fillna(3).astype(int)

print("Distribusi skor C1 pada grid:")
print(grid["skor_c1"].value_counts().sort_index())
```

```python
# ============================================================
# C2. DEMAND GAP — RADIUS 3 KM PER SEL GRID (BUKAN kecamatan)
# ============================================================
# Masalah lama: C2 per kecamatan -> seluruh poligon satu warna.
# Solusi: setiap sel 500m dihitung sendiri berdasarkan:
#   total siswa dalam radius 3km  minus  kapasitas SPPG dalam radius 3km
# Hasilnya: demand gap yang benar-benar terlokalisasi.

RADIUS_DEMAND = 3000   # meter
KAPASITAS_SPPG = 2000  # porsi/hari per SPPG

grid_coords_arr = grid[["cx", "cy"]].values

# Array sekolah + siswa
sekolah_coords_arr = np.array([(g.x, g.y) for g in sekolah_m.geometry])
sekolah_siswa_arr  = sekolah_m["jumlah_siswa"].fillna(0).values

# Array SPPG existing
sppg_coords_arr = np.array([(g.x, g.y) for g in sppg_m.geometry])

tree_sekolah = cKDTree(sekolah_coords_arr)
tree_sppg    = cKDTree(sppg_coords_arr)

siswa_3km     = []
kapasitas_3km = []

for coord in grid_coords_arr:
    idx_s    = tree_sekolah.query_ball_point(coord, RADIUS_DEMAND)
    idx_sppg = tree_sppg.query_ball_point(coord, RADIUS_DEMAND)
    siswa_3km.append(float(sekolah_siswa_arr[idx_s].sum()) if len(idx_s) > 0 else 0.0)
    kapasitas_3km.append(len(idx_sppg) * KAPASITAS_SPPG)

grid["total_siswa_3km"]    = siswa_3km
grid["kapasitas_sppg_3km"] = kapasitas_3km
grid["demand_gap_3km"] = (
    np.array(siswa_3km) - np.array(kapasitas_3km)
).clip(min=0)

def skor_demand_gap_grid(gap):
    if gap == 0:
        return 1
    elif gap <= 2000:
        return 2
    elif gap <= 5000:
        return 3
    elif gap <= 10000:
        return 4
    else:
        return 5

grid["skor_c2"] = grid["demand_gap_3km"].apply(skor_demand_gap_grid)

print("Distribusi skor C2 (demand gap berbasis radius 3km):")
print(grid["skor_c2"].value_counts().sort_index())
display(grid[["total_siswa_3km", "kapasitas_sppg_3km", "demand_gap_3km"]].describe())
```

```python
# ============================================================
# VISUALISASI C2 — DEMAND GAP RADIUS 3 KM (GRID)
# ============================================================

fig, ax = plt.subplots(figsize=(12, 12))

surabaya_m.plot(ax=ax, color="#f5f5f5", edgecolor="none", zorder=1)

cmap_c2 = ListedColormap(["#ffffcc", "#a1dab4", "#41b6c4", "#2c7fb8", "#253494"])

grid.plot(
    ax=ax, column="skor_c2", cmap=cmap_c2,
    edgecolor="none", alpha=0.85, zorder=2
)

surabaya_m.boundary.plot(ax=ax, color="black", linewidth=0.8, zorder=3)
sppg_m.plot(ax=ax, markersize=50, marker="*",
            color="yellow", edgecolor="black", linewidth=0.5, zorder=5)

legend_items = [
    mpatches.Patch(color="#ffffcc", label="Skor 1: Kapasitas cukup / surplus"),
    mpatches.Patch(color="#a1dab4", label="Skor 2: Gap <= 2.000 siswa"),
    mpatches.Patch(color="#41b6c4", label="Skor 3: Gap 2.001-5.000 siswa"),
    mpatches.Patch(color="#2c7fb8", label="Skor 4: Gap 5.001-10.000 siswa"),
    mpatches.Patch(color="#253494", label="Skor 5: Gap > 10.000 siswa"),
]
ax.legend(handles=legend_items, title="C2 Demand Gap (radius 3km)",
          loc="upper right", frameon=True, fontsize=9)
ax.set_aspect("equal")
plt.title("C2 - Demand Gap Siswa dalam Radius 3km per Sel 500m\nBukan per Kecamatan", fontsize=12)
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# C3. KEPADATAN SEKOLAH — KERNEL DENSITY ESTIMATION (GRID)
# ============================================================
# Masalah lama: C3 = jumlah sekolah per kecamatan
#   -> double counting dengan C2 (banyak sekolah ~ banyak siswa)
#   -> seluruh kecamatan satu warna
# Solusi: KDE menghasilkan kepadatan titik sekolah per lokasi.
# Wilayah tanpa sekolah otomatis skor rendah.
# Hotspot sekolah skor tinggi. Tidak ada lagi double counting.

sekolah_x = sekolah_m.geometry.x.values
sekolah_y = sekolah_m.geometry.y.values

kde = gaussian_kde(
    np.vstack([sekolah_x, sekolah_y]),
    bw_method=0.15   # bandwidth relatif terhadap IQR data
)

grid["kde_sekolah"] = kde(np.vstack([grid["cx"].values, grid["cy"].values]))

grid["skor_c3"] = pd.qcut(
    grid["kde_sekolah"],
    q=5,
    labels=[1, 2, 3, 4, 5],
    duplicates="drop"
).astype(int)

print("Distribusi skor C3 (KDE kepadatan sekolah):")
print(grid["skor_c3"].value_counts().sort_index())
```

```python
# ============================================================
# VISUALISASI C3 — KDE KEPADATAN SEKOLAH (GRID)
# ============================================================

fig, ax = plt.subplots(figsize=(12, 12))

surabaya_m.plot(ax=ax, color="#f5f5f5", edgecolor="none", zorder=1)

cmap_c3 = ListedColormap(["#edf8fb", "#b2e2e2", "#66c2a4", "#2ca25f", "#006d2c"])

grid.plot(
    ax=ax, column="skor_c3", cmap=cmap_c3,
    edgecolor="none", alpha=0.85, zorder=2
)

surabaya_m.boundary.plot(ax=ax, color="black", linewidth=0.8, zorder=3)
sekolah_m.plot(ax=ax, markersize=2, color="black", alpha=0.4, zorder=4)
sppg_m.plot(ax=ax, markersize=50, marker="*",
            color="red", edgecolor="black", linewidth=0.5, zorder=6)

legend_items = [
    mpatches.Patch(color="#edf8fb", label="Skor 1: Kepadatan sangat rendah"),
    mpatches.Patch(color="#b2e2e2", label="Skor 2: Kepadatan rendah"),
    mpatches.Patch(color="#66c2a4", label="Skor 3: Kepadatan sedang"),
    mpatches.Patch(color="#2ca25f", label="Skor 4: Kepadatan tinggi"),
    mpatches.Patch(color="#006d2c", label="Skor 5: Hotspot sekolah"),
]
ax.legend(handles=legend_items, title="C3 KDE Kepadatan Sekolah",
          loc="upper right", frameon=True, fontsize=9)
ax.set_aspect("equal")
plt.title("C3 - Kernel Density Kepadatan Sekolah per Sel 500m\nTitik hitam = lokasi sekolah aktual", fontsize=12)
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# C4. RASIO LAYANAN SPPG — KECAMATAN -> JOIN KE GRID
# ============================================================
# Fungsi: konteks makro (apakah kecamatan ini secara umum
# sudah terlayani). Tetap per kecamatan tapi bukan unit utama.

sekolah_kec_join = gpd.sjoin(
    sekolah_m,
    surabaya_m[["Kecamatan", "geometry"]],
    how="left", predicate="within"
)
demand_kec = (
    sekolah_kec_join
    .groupby("Kecamatan")
    .agg(jumlah_sekolah=("nama", "count"),
         total_siswa=("jumlah_siswa", "sum"))
    .reset_index()
)

sppg_kec_join = gpd.sjoin(
    sppg_m,
    surabaya_m[["Kecamatan", "geometry"]],
    how="left", predicate="within"
)
jml_sppg_kec = (
    sppg_kec_join
    .groupby("Kecamatan")
    .agg(jumlah_sppg=("nama", "count"))
    .reset_index()
)

KAPASITAS_SPPG = 2000
c4_kec = surabaya_m.merge(demand_kec, on="Kecamatan", how="left")
c4_kec = c4_kec.merge(jml_sppg_kec, on="Kecamatan", how="left")
c4_kec[["total_siswa", "jumlah_sppg"]] = (
    c4_kec[["total_siswa", "jumlah_sppg"]].fillna(0)
)
c4_kec["kapasitas_total"] = c4_kec["jumlah_sppg"] * KAPASITAS_SPPG
c4_kec["rasio_layanan"] = (
    c4_kec["kapasitas_total"] /
    c4_kec["total_siswa"].replace(0, np.nan)
).fillna(0).replace([np.inf, -np.inf], 0)

def skor_rasio_c4(rasio):
    if rasio >= 1.00:   return 1
    elif rasio >= 0.75: return 2
    elif rasio >= 0.50: return 3
    elif rasio >= 0.25: return 4
    else:               return 5

c4_kec["skor_c4"] = c4_kec["rasio_layanan"].apply(skor_rasio_c4)

# Join ke grid
g_c4 = gpd.sjoin(
    grid[["grid_id", "geometry"]],
    c4_kec[["Kecamatan", "skor_c4", "geometry"]],
    how="left", predicate="intersects"
)
g_c4 = g_c4.groupby("grid_id")["skor_c4"].first().reset_index()
grid = grid.merge(g_c4, on="grid_id", how="left")
grid["skor_c4"] = grid["skor_c4"].fillna(3).astype(int)

print("Distribusi skor C4:")
print(grid["skor_c4"].value_counts().sort_index())
display(c4_kec[["Kecamatan", "total_siswa", "jumlah_sppg", "rasio_layanan", "skor_c4"]]
        .sort_values("skor_c4", ascending=False))
```

```python

# C5. AKSES JALAN — KECAMATAN -> JOIN KE GRID


def skor_kelas_jalan(kelas):
    if kelas in ["Jalan Arteri", "Jalan Nasional"]: return 5
    elif kelas == "Jalan Kolektor":                  return 4
    elif kelas == "Jalan Lokal":                     return 3
    elif kelas == "Jalan Lain":                      return 2
    else:                                            return 1

jalan_c5 = jalan_m.copy()
jalan_c5["skor_jalan"] = jalan_c5["REMARK"].apply(skor_kelas_jalan)

jalan_kec = gpd.overlay(
    jalan_c5[["REMARK", "skor_jalan", "geometry"]],
    surabaya_m[["Kecamatan", "geometry"]],
    how="intersection"
)
jalan_kec["panjang_m"]   = jalan_kec.geometry.length
jalan_kec["nilai_jalan"] = jalan_kec["skor_jalan"] * jalan_kec["panjang_m"]

skor_jalan_kec = (
    jalan_kec
    .groupby("Kecamatan")
    .agg(total_panjang=("panjang_m", "sum"),
         total_nilai=("nilai_jalan", "sum"))
    .reset_index()
)
skor_jalan_kec["skor_c5_raw"] = (
    skor_jalan_kec["total_nilai"] / skor_jalan_kec["total_panjang"]
)
skor_jalan_kec["skor_c5"] = pd.qcut(
    skor_jalan_kec["skor_c5_raw"], q=5,
    labels=[1, 2, 3, 4, 5], duplicates="drop"
).astype(int)

c5_kec = surabaya_m.merge(skor_jalan_kec, on="Kecamatan", how="left")
c5_kec["skor_c5"] = c5_kec["skor_c5"].fillna(1).astype(int)

# Join ke grid
g_c5 = gpd.sjoin(
    grid[["grid_id", "geometry"]],
    c5_kec[["Kecamatan", "skor_c5", "geometry"]],
    how="left", predicate="intersects"
)
g_c5 = g_c5.groupby("grid_id")["skor_c5"].first().reset_index()
grid = grid.merge(g_c5, on="grid_id", how="left")
grid["skor_c5"] = grid["skor_c5"].fillna(1).astype(int)

print("Distribusi skor C5:")
print(grid["skor_c5"].value_counts().sort_index())
```

```python

# REKAP SKOR + HITUNG AHP PER SEL GRID

# Bobot AHP (direvisi):
#   C1 Jangkauan gap SPPG        0.35  PRIMARY: lokasi gap
#   C2 Demand gap radius 3km     0.30  PRIMARY: besaran unmet demand
#   C3 KDE kepadatan sekolah     0.20  SECONDARY: konsentrasi spasial
#   C4 Rasio layanan kecamatan   0.10  KONTEKS MAKRO
#   C5 Akses jalan               0.05  ENABLING
#   Total                        1.00
#
# Perubahan dari model lama:
#   C1 naik (0.25->0.35) karena paling langsung ukur gap
#   C2 tetap bobot tinggi tapi kini berbasis radius, bukan kecamatan
#   C3 naik tipis (0.20) tapi kini KDE, tidak double count C2

print("Rekap skor semua parameter per grid:")
display(grid[["skor_c1","skor_c2","skor_c3","skor_c4","skor_c5"]].describe())

W = {"c1": 0.35, "c2": 0.30, "c3": 0.20, "c4": 0.10, "c5": 0.05}

grid["skor_ahp"] = (
    grid["skor_c1"] * W["c1"] +
    grid["skor_c2"] * W["c2"] +
    grid["skor_c3"] * W["c3"] +
    grid["skor_c4"] * W["c4"] +
    grid["skor_c5"] * W["c5"]
)

print("\nStatistik skor AHP per grid:")
display(grid["skor_ahp"].describe())
print(f"Range aktual: {grid['skor_ahp'].min():.3f} - {grid['skor_ahp'].max():.3f}")
```

```python

# CONSTRAINT 1. DALAM RADIUS 750 M DARI SEKOLAH

# Direvisi: 1.500m -> 750m
# Alasan: Surabaya sangat padat sekolah. Buffer 1.500m
# menyebabkan hampir seluruh kota lolos constraint sehingga
# tidak selektif. 750m = jarak operasional realistis
# dari SPPG menuju sekolah di kota padat.

sekolah_buf_750 = sekolah_m.copy()
sekolah_buf_750["geometry"] = sekolah_buf_750.buffer(750)
constraint_sekolah = sekolah_buf_750.dissolve()
constraint_sekolah = gpd.clip(constraint_sekolah, surabaya_union)

luas_c = constraint_sekolah.geometry.area.sum() / 1e6
luas_s = surabaya_union.geometry.area.sum() / 1e6
print(f"Luas dalam 750m sekolah : {luas_c:.2f} km2")
print(f"Luas Surabaya           : {luas_s:.2f} km2")
print(f"Persentase              : {luas_c/luas_s*100:.1f}%")
```

```python

# CONSTRAINT 2. BUANG BADAN JALAN (BUFFER 15m)
# + APPLY SEMUA CONSTRAINT KE GRID

jalan_buf_15 = jalan_m.copy()
jalan_buf_15["geometry"] = jalan_buf_15.buffer(15)
jalan_union_geom = jalan_buf_15.dissolve().geometry.iloc[0]

# 1. Clip grid ke dalam 750m sekolah
grid_constrained = gpd.clip(grid, constraint_sekolah)
grid_constrained = grid_constrained.reset_index(drop=True)

# 2. Kurangi badan jalan dari tiap sel
grid_constrained["geometry"] = grid_constrained.geometry.difference(
    jalan_union_geom
)

# 3. Buang geometri kosong atau sangat kecil (< 500 m2)
grid_constrained = grid_constrained[~grid_constrained.is_empty]
grid_constrained = grid_constrained[grid_constrained.geometry.area > 500]
grid_constrained = grid_constrained.reset_index(drop=True)

print(f"Grid sebelum constraint : {len(grid):,} sel")
print(f"Grid setelah constraint : {len(grid_constrained):,} sel")
print(f"Yang lolos              : {len(grid_constrained)/len(grid)*100:.1f}%")
```

```python

# KLASIFIKASI AHP FINAL (GRID LOLOS CONSTRAINT)


def kelas_ahp(skor):
    if skor < 2.0:   return "Tidak Prioritas"
    elif skor < 2.8: return "Prioritas Rendah"
    elif skor < 3.6: return "Prioritas Sedang"
    elif skor < 4.3: return "Prioritas Tinggi"
    else:            return "Sangat Prioritas"

urutan_kelas = {
    "Tidak Prioritas": 1,
    "Prioritas Rendah": 2,
    "Prioritas Sedang": 3,
    "Prioritas Tinggi": 4,
    "Sangat Prioritas": 5,
}

grid_constrained["kelas_ahp"]  = grid_constrained["skor_ahp"].apply(kelas_ahp)
grid_constrained["kode_kelas"] = grid_constrained["kelas_ahp"].map(urutan_kelas)

print("Distribusi kelas AHP (lolos constraint):")
print(grid_constrained["kelas_ahp"].value_counts())
print()
for k in ["Sangat Prioritas", "Prioritas Tinggi", "Prioritas Sedang"]:
    luas = grid_constrained[grid_constrained["kelas_ahp"]==k].geometry.area.sum()/1e6
    n    = len(grid_constrained[grid_constrained["kelas_ahp"]==k])
    print(f"  {k}: {luas:.2f} km2 ({n} sel)")
```

```python
# ============================================================
# VISUALISASI AHP FINAL — SEMUA KELAS (GRID 500m)
# ============================================================

fig, ax = plt.subplots(figsize=(13, 13))

surabaya_m.plot(ax=ax, color="#f4f4f0", edgecolor="none", zorder=1)

# Palette muted sequential: abu -> hijau sage -> biru medium -> biru tua -> navy
cmap_ahp = ListedColormap([
    "#e8e8e4",  # 1 Tidak Prioritas
    "#c9dcc5",  # 2 Prioritas Rendah  (sage muted)
    "#7db3c4",  # 3 Prioritas Sedang  (steel blue muted)
    "#3d7da8",  # 4 Prioritas Tinggi  (biru medium)
    "#1b4f72"   # 5 Sangat Prioritas  (navy tua)
])

grid_constrained.plot(
    ax=ax, column="kode_kelas", cmap=cmap_ahp,
    edgecolor="none", alpha=0.88, zorder=2
)

surabaya_m.boundary.plot(ax=ax, color="#aaaaaa", linewidth=0.3, zorder=3)
surabaya_m.boundary.plot(ax=ax, color="#444444", linewidth=1.0, zorder=6)
sekolah_m.plot(ax=ax, markersize=2, color="#555555", alpha=0.30, zorder=4)
sppg_m.plot(ax=ax, markersize=65, marker="*",
            color="#e8c96a", edgecolor="#6b5700", linewidth=0.7, zorder=7)

legend_items = [
    mpatches.Patch(color="#e8e8e4", label="Tidak Prioritas"),
    mpatches.Patch(color="#c9dcc5", label="Prioritas Rendah"),
    mpatches.Patch(color="#7db3c4", label="Prioritas Sedang"),
    mpatches.Patch(color="#3d7da8", label="Prioritas Tinggi"),
    mpatches.Patch(color="#1b4f72", label="Sangat Prioritas"),
    Line2D([0],[0], marker="*", color="w", markerfacecolor="#e8c96a",
           markeredgecolor="#6b5700", markersize=12, label="SPPG Existing"),
    Line2D([0],[0], marker="o", color="w", markerfacecolor="#555555",
           markersize=5, alpha=0.4, label="Sekolah SD/SMP/SMA"),
]
ax.legend(handles=legend_items, title="Tingkat Prioritas",
          loc="upper right", frameon=True, fontsize=9,
          fancybox=False, edgecolor="#cccccc")
ax.set_aspect("equal")
plt.title("Peta AHP Prioritas SPPG — Grid 500m + Constraint 750m\nKota Surabaya",
          fontsize=13, fontweight="bold", color="#222222")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# ANALISIS KEBUTUHAN SPPG BARU PER KECAMATAN
# ============================================================
# Logika:
#   demand_gap_kecamatan = total_siswa - (jumlah_sppg_existing * 2000)
#   kebutuhan_sppg_baru  = ceil(demand_gap / 2000)
#   Hanya kecamatan yang memiliki kawasan rekomendasi yang dihitung.

# 1. Identifikasi kecamatan yang punya kawasan rekomendasi
rek_kec_join = gpd.sjoin(
    rekomendasi[["grid_id", "skor_ahp", "kelas_ahp", "geometry"]],
    surabaya_m[["Kecamatan", "geometry"]],
    how="left",
    predicate="intersects"
)

kec_rek_agg = (
    rek_kec_join
    .groupby("Kecamatan")
    .agg(
        jumlah_sel_rek=("grid_id", "count"),
        luas_rek_km2=("geometry", lambda x: x.area.sum() / 1e6),
        skor_ahp_max=("skor_ahp", "max"),
        skor_ahp_mean=("skor_ahp", "mean"),
        pct_sangat=("kelas_ahp", lambda x: round((x == "Sangat Prioritas").sum() / len(x) * 100, 1))
    )
    .reset_index()
)

# 2. Join dengan data demand & SPPG per kecamatan
kec_rek_agg = kec_rek_agg.merge(
    c4_kec[["Kecamatan", "total_siswa", "jumlah_sppg", "kapasitas_total", "rasio_layanan"]],
    on="Kecamatan", how="left"
)

KAPASITAS_SPPG = 2000
kec_rek_agg["demand_gap_kec"] = (
    kec_rek_agg["total_siswa"] - kec_rek_agg["kapasitas_total"]
).clip(lower=0)
kec_rek_agg["kebutuhan_sppg_baru"] = np.ceil(
    kec_rek_agg["demand_gap_kec"] / KAPASITAS_SPPG
).astype(int)

kec_rek_agg = kec_rek_agg.sort_values("kebutuhan_sppg_baru", ascending=True)

# ── Visualisasi: Horizontal bar chart ──────────────────────

fig, ax = plt.subplots(figsize=(13, max(7, len(kec_rek_agg) * 0.45)))

# Warna bar berdasarkan skor_ahp_max (muted sienna)
norm = plt.Normalize(kec_rek_agg["skor_ahp_max"].min(),
                     kec_rek_agg["skor_ahp_max"].max())
colors_bar = [
    plt.cm.get_cmap("YlOrBr")(norm(v)) for v in kec_rek_agg["skor_ahp_max"]
]

bars = ax.barh(
    kec_rek_agg["Kecamatan"],
    kec_rek_agg["kebutuhan_sppg_baru"],
    color=colors_bar,
    edgecolor="white",
    linewidth=0.6,
    height=0.7
)

# Annotasi: angka kebutuhan + existing
for i, (_, row) in enumerate(kec_rek_agg.iterrows()):
    n   = int(row["kebutuhan_sppg_baru"])
    eks = int(row["jumlah_sppg"])
    ax.text(n + 0.08, i,
            f"+{n} baru  (existing: {eks})",
            va="center", ha="left", fontsize=8.5, color="#333333")

# Grid & styling
ax.set_xlabel("Estimasi Jumlah SPPG Baru yang Dibutuhkan", fontsize=11, color="#333333")
ax.set_title(
    "Estimasi Kebutuhan SPPG Baru per Kecamatan\n"
    "(hanya kecamatan yang memiliki kawasan rekomendasi, kapasitas 1 SPPG = 2.000 siswa)",
    fontsize=12, fontweight="bold", pad=12, color="#222222"
)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#cccccc")
ax.spines["bottom"].set_color("#cccccc")
ax.tick_params(axis="y", labelsize=9, colors="#333333")
ax.tick_params(axis="x", labelsize=9, colors="#555555")
ax.set_xlim(0, kec_rek_agg["kebutuhan_sppg_baru"].max() + 6)
ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.grid(axis="x", linestyle="--", linewidth=0.5, alpha=0.5, color="#bbbbbb")

# Colorbar
sm = plt.cm.ScalarMappable(cmap="YlOrBr", norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, orientation="vertical", fraction=0.015, pad=0.02)
cbar.set_label("Skor AHP Maks\n(kawasan rekomendasi)", fontsize=8, color="#333333")
cbar.ax.tick_params(labelsize=7)

plt.tight_layout()
plt.savefig("../output/analisis_kebutuhan_sppg_kecamatan.png",
            dpi=200, bbox_inches="tight", facecolor="white")
plt.show()

print("\n=== TABEL KEBUTUHAN SPPG PER KECAMATAN ===")
display(
    kec_rek_agg[["Kecamatan", "total_siswa", "jumlah_sppg",
                 "demand_gap_kec", "kebutuhan_sppg_baru",
                 "luas_rek_km2", "skor_ahp_max"]]
    .sort_values("kebutuhan_sppg_baru", ascending=False)
    .rename(columns={
        "total_siswa":          "Siswa",
        "jumlah_sppg":          "SPPG Existing",
        "demand_gap_kec":       "Demand Gap",
        "kebutuhan_sppg_baru":  "SPPG Baru",
        "luas_rek_km2":         "Luas Rek (km2)",
        "skor_ahp_max":         "Skor AHP Maks"
    })
    .reset_index(drop=True)
)

total_baru = kec_rek_agg["kebutuhan_sppg_baru"].sum()
print(f"\n  Total estimasi SPPG baru yang dibutuhkan : {total_baru}")
print(f"  Total SPPG existing seluruh Surabaya    : {len(sppg_m)}")
print(f"  Kecamatan yang memiliki kawasan rek.    : {len(kec_rek_agg)}")
```
